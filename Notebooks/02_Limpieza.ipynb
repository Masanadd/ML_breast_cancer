## **2.1. Descripción de las Variables**

A continuación, una breve descripción de cada una de estas variables basándome en la terminología médica y los contextos comunes en estudios de cáncer de mama. Estas definiciones pueden variar ligeramente dependiendo del sistema de registro utilizado, pero en general representan las siguientes características:

- **Patient ID:** Un identificador único para cada paciente.
- **Age at Diagnosis:** La edad del paciente en el momento del diagnóstico.
- **Type of Breast Surgery:** El tipo de cirugía mamaria que recibió el paciente (por ejemplo, mastectomía, tumorectomía).
- **Cancer Type:** El tipo general de cáncer (como carcinoma ductal, carcinoma lobulillar).
- **Cancer Type Detailed:** Una descripción más específica del tipo de cáncer.
- **Cellularity:** El grado de celularidad del tumor, que indica la densidad de células cancerígenas presentes.
- **Chemotherapy:** Indica si el paciente recibió quimioterapia como parte de su tratamiento.
- **Pam50 + Claudin-low subtype:** Un subgrupo molecular del cáncer basado en la clasificación PAM50, que incluye información sobre la expresión génica y la clasificación de subtipos como luminal A, luminal B, HER2-enriquecido, entre otros.
- **Cohort:** El grupo o cohorte al que pertenece el paciente dentro del estudio.
- **ER status measured by IHC:** Estado del receptor de estrógeno medido por inmunohistoquímica (IHC), que puede ser positivo o negativo.
- **ER Status:** Simplificación del estado del receptor de estrógeno (positivo o negativo).
- **Neoplasm Histologic Grade:** El grado histológico del tumor, que indica qué tan anormales se ven las células cancerosas bajo el microscopio y qué tan probable es que crezcan y se diseminen.
- **HER2 status measured by SNP6:** Estado del receptor HER2 medido mediante análisis genético, como el de un microarray de SNP.
- **HER2 Status:** Simplificación del estado HER2 (positivo, negativo o indeterminado).
- **Tumor Other Histologic Subtype:** Subtipo histológico adicional del tumor, que proporciona detalles más granulares sobre la clasificación histológica.
- **Hormone Therapy:** Indica si el paciente recibió terapia hormonal como parte del tratamiento.
- **Inferred Menopausal State:** Estado menopáusico inferido del paciente (por ejemplo, premenopáusica o postmenopáusica).
- **Integrative Cluster:** Una clasificación molecular que agrupa tumores con características similares basadas en múltiples tipos de datos moleculares.
- **Primary Tumor Laterality:** La lateralidad del tumor primario, indicando si el tumor estaba en el seno izquierdo o derecho.
- **Lymph nodes examined positive:** Número de ganglios linfáticos examinados que resultaron positivos para células cancerígenas.
- **Mutation Count:** Número total de mutaciones identificadas en el análisis genético del tumor.
- **Nottingham prognostic index:** Un índice pronóstico calculado en función del tamaño del tumor, el grado histológico y el estado de los ganglios linfáticos, usado para predecir el resultado del cáncer de mama.
- **Oncotree Code:** Código del árbol oncológico que clasifica los cánceres en subtipos y categorías con base en características moleculares y clínicas.
- **Overall Survival (Months):** Tiempo total de supervivencia en meses desde el diagnóstico o inicio del tratamiento hasta la muerte o la última revisión.
- **Overall Survival Status:** Estado de supervivencia general (si el paciente sigue vivo o ha fallecido).
- **PR Status:** Estado del receptor de progesterona (positivo o negativo).
- **Radio Therapy:** Indica si el paciente recibió radioterapia como parte del tratamiento.
- **Relapse Free Status (Months):** Tiempo en meses en que el paciente permaneció libre de recaídas después del tratamiento inicial.
- **Relapse Free Status:** Indica si el paciente ha experimentado una recaída.
- **Sex:** Sexo del paciente (masculino, femenino u otro).
- **3-Gene classifier subtype:** Un subgrupo del cáncer basado en la expresión de tres genes específicos, utilizado para clasificar tumores.
- **Tumor Size:** Tamaño del tumor primario, generalmente en milímetros.
- **Tumor Stage:** Etapa del cáncer, que describe la extensión de la enfermedad (por ejemplo, etapa I, II, III o IV).
- **Patient's Vital Status:** Estado vital del paciente al momento del análisis (vivo o fallecido).

Estas definiciones te permiten comprender qué tipo de información representa cada variable y cómo podrían relacionarse con el análisis clínico y pronóstico del cáncer de mama.

Después de analizar la correlación, usaremos las siguientes estrategias:

| **Variable**                          | **Método de Imputación**                                      | **Justificación** |
|--------------------------------------|--------------------------------------------------|----------------|
| `Age at Diagnosis`                  | Mediana por **subtipo de cáncer**               | La edad de diagnóstico varía por tipo de tumor. |

## - Decisión final sobre estrategia de imputación:

1. **Imputación de "Age at Diagnosis" usando la mediana por subtipo de cáncer**

2. **Para "Lymph nodes examined positive" y "Nottingham prognostic index"**:
   - Primero imputamos valores nulos con la **mediana**, luego aplicamos regresión.

3. **Imputación de valores nulos en variable Mutation Count con la mediana"**

4. **Para "Tumor Stage" y "Tumor Size"**:
   - Imputamos **"Tumor Stage"** basado en el tamaño tumoral.

5. **Imputar "Overall Survival (Months)" con la mediana dentro de "Relapse Free Status (Months)"**




🔹Hacemos copia de df_clean a df_imputed antes de imputar las variables.

```python
from sklearn.impute import SimpleImputer
```

```python
df_imputed=df_clean.copy()
```

Imputación de "Age at Diagnosis" usando la mediana por subtipo de cáncer

```python
# Verificar si la variable "Cancer Type Detailed" está en el dataset
if "Cancer Type Detailed" in df_imputed.columns:
    # Imputar valores faltantes usando la mediana de la edad por subtipo de cáncer
    df_imputed["Age at Diagnosis"] = df_imputed.groupby("Cancer Type Detailed")["Age at Diagnosis"].transform(lambda x: x.fillna(x.median()))
else:
    print("⚠️ La variable 'Cancer Type Detailed' no está en el dataset.")

# Verificar si quedan valores nulos en "Age at Diagnosis"
df_imputed["Age at Diagnosis"].isnull().sum()
```

Imputación de valores faltatnes en dos columnas "Lymph nodes examined positive" y "Nottingham prognostic index" con la mediana antes de la regresión

```python
imputer_median = SimpleImputer(strategy="median")
df_imputed["Lymph nodes examined positive"] = imputer_median.fit_transform(df_imputed[["Lymph nodes examined positive"]])
df_imputed["Nottingham prognostic index"] = imputer_median.fit_transform(df_imputed[["Nottingham prognostic index"]])
print(df_imputed["Lymph nodes examined positive"].isnull().sum())
print(df_imputed["Nottingham prognostic index"].isnull().sum())
```

Imputación de "Mutation Count" usando la mediana

```python
df_imputed["Mutation Count"] = df_imputed["Mutation Count"].fillna(df_imputed["Mutation Count"].median())
df_imputed["Mutation Count"].isnull().sum()
```

Imputación de "Tumor Stage" usando la función ya que conocemos las reglas clínicas que relacionan ambas varibles timor size y tumor stage.

```python
def impute_tumor_stage(row):
    if pd.isna(row["Tumor Stage"]):  
        if row["Tumor Size"] <= 20:  
            return 1  
        elif row["Tumor Size"] <= 50:  
            return 2  
        else:  
            return 3  
    else:
        return row["Tumor Stage"]  

# Aplicar la función a cada fila
df_imputed["Tumor Stage"] = df_imputed.apply(impute_tumor_stage, axis=1)
df_imputed["Tumor Stage"].isnull().sum()
```

Imputación de "Tumor Size" con la mediana dentro de "Tumor Stage".

```python
df_imputed["Tumor Size"] = df_imputed.groupby("Tumor Stage")["Tumor Size"].transform(lambda x: x.fillna(x.median()))
df_imputed["Tumor Size"].isnull().sum()
```

 Imputar "Overall Survival (Months)" con la mediana dentro de relapse free status months

```python
df_imputed["Overall Survival (Months)"] = (
    df_imputed.groupby("Relapse Free Status (Months)")["Overall Survival (Months)"]
    .transform(lambda x: x.fillna(x.median())  # Imputar con la mediana del grupo
))
missing_overall_survival=df_imputed["Overall Survival (Months)"].isnull().sum()
```

 Si aún hay valores nulos, imputar con la mediana global del dataset

```python
if missing_overall_survival > 0:
    df_imputed["Overall Survival (Months)"] = df_imputed["Overall Survival (Months)"].fillna(
        df_imputed["Overall Survival (Months)"].median()
    )
    
df_imputed["Overall Survival (Months)"].isnull().sum()
```

## **Imputación de variables categóricas basados en relaciones clínicas**

### **Estrategias de Imputación por Variable**

1. **Variables de Tratamiento**:
   - **"Chemotherapy"**, **"Hormone Therapy"**, **"Radio Therapy"**:
     - Imputar usando la **moda** dentro de cada subtipo de cáncer (`Cancer Type Detailed`).

2. **Variables Relacionadas con el Cáncer**:
   - **"Type of Breast Surgery"**:
     - Imputar por **moda** basada en `"Cancer Type Detailed"`.
   - **"Cellularity"**:
     - Imputar por **moda** dentro de `"Cancer Type Detailed"`.
   - **"Neoplasm Histologic Grade"**:
     - Imputar por **mediana** dentro de `"Tumor Stage"`.
   - **"Tumor Other Histologic Subtype"**:
     - Imputar por **moda** dentro de `"Cancer Type Detailed"`.

3. **Variables Moleculares y Biomarcadores**:
   - **"Pam50 + Claudin-low subtype"**, **"HER2 Status"**, **"PR Status"**, **"ER Status"**:
     - Imputar usando la **moda** en función de `"Cancer Type Detailed"` y otros biomarcadores relacionados (`ER Status`, `PR Status`).
   - **"HER2 status measured by SNP6"**:
     - Imputar basado en `"ER Status"`, `"PR Status"` y `"Age at Diagnosis"`.
   - **"ER status measured by IHC"**:
     - Imputar usando la **moda** basada en `"ER Status"`.

4. **Variables Clínicas y Demográficas**:
   - **"Inferred Menopausal State"**:
     - Imputar basado en `"ER Status"`, `"PR Status"` y `"Age at Diagnosis"`.
     - **"Patient's Vital Status"**:
     - Asignar basado en `"Overall Survival Status"`.

5. **Variables con Asignación Directa**:
   - **"3-Gene classifier subtype"**:
     - Asignar `"Unknown"` a los valores nulos.
   - **"Integrative Cluster"**:
     - Asignar `"Unknown"`.


Variables de tratamiento con la moda según subtipo de cáncer

```python
categorical_treatment_vars = ["Chemotherapy", "Hormone Therapy", "Radio Therapy"]
for var in categorical_treatment_vars:
    df_imputed[var] = df_imputed.groupby("Cancer Type Detailed")[var].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "No"))
```

```python
df_imputed["Neoplasm Histologic Grade"] = df_imputed.groupby("Tumor Size")["Neoplasm Histologic Grade"].transform(
    lambda x: x.fillna(x.median()) if x.notna().any() else x 
)
df_imputed["Tumor Other Histologic Subtype"] = df_imputed.groupby("Cancer Type Detailed")["Tumor Other Histologic Subtype"].transform(
    lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x  
)

categorical_vars = ["Type of Breast Surgery", "Cellularity"]
for var in categorical_vars:
    df_imputed[var] = df_imputed.groupby("Cancer Type Detailed")[var].transform(
        lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else "Unknown"  
    )
```

```python
for col in ["Pam50 + Claudin-low subtype", "HER2 Status", "PR Status", "ER Status"]:
    df_imputed[col] = df_imputed.groupby("Cancer Type Detailed")[col].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown"))

df_imputed["HER2 status measured by SNP6"] = df_imputed.groupby(["ER Status", "PR Status", "Age at Diagnosis"])["HER2 status measured by SNP6"].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown")
)

df_imputed["ER status measured by IHC"] = df_imputed.groupby("ER Status")["ER status measured by IHC"].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown")
)
```

```python
print(df_imputed["Patient's Vital Status"].value_counts(dropna=False))
```

```python
df_imputed["Inferred Menopausal State"] = df_imputed.groupby(["ER Status", "PR Status", "Age at Diagnosis"])["Inferred Menopausal State"].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown")
)
df_imputed["Patient's Vital Status"] = df_imputed["Overall Survival Status"].map({"Living": "Alive", "Deceased": "Dead"})
```

Imputación de Overall Survival Status basada en "Patient's Vital Status", se usó "Patient's Vital Status" para inferir si el paciente estaba "Living" o "Deceased"

```python
if "3-Gene classifier subtype" in df_imputed.columns:
    df_imputed["3-Gene classifier subtype"] =df_imputed["3-Gene classifier subtype"].fillna("Unknown")

if "Integrative Cluster" in df_imputed.columns:
    df_imputed["Integrative Cluster"] = df_imputed["Integrative Cluster"].fillna("Unknown")
```

```python
df_imputed.loc[
    (df_imputed["Overall Survival Status"].isna()) & (df_imputed["Overall Survival (Months)"] < 12),
    "Overall Survival Status"
] = "Deceased"

df_imputed.loc[
    (df_imputed["Overall Survival Status"].isna()) & (df_imputed["Overall Survival (Months)"] >= 60),
    "Overall Survival Status"
] = "Living"
```

Se usó la duración de la supervivencia (Overall survival Status (months)) para imputar valores faltantes. Pacientes con menos de 12 meses de supervivencia se imputaron como "Deceased", y aquellos con más de 60 meses como "Living".

## **Agrupación de Variables Categóricas Poco Frecuentes**

Para variables categóricas con muchas clases poco frecuentes, podemos agruparlas en una categoría general llamada **"Otros"**.

#### **1. Cancer Type Detailed**
- Contiene subtipos de cáncer de mama como **Ductal Carcinoma, Lobular Carcinoma, Medullary, etc.**  
- Algunos subtipos tienen **muy pocas observaciones**.  
- **Solución**: Mantener los **2-4 subtipos más comunes** y agrupar el resto en **"Otros"**.  

#### **2. Pam50 + Claudin-low Subtype**
- Contiene subtipos moleculares como **Luminal A, Luminal B, Basal, etc.**  
- Algunos subtipos como **"NC"** o **"Normal"** tienen **muy pocas muestras**.  
- **Solución**: Mantener **Luminal A, Luminal B, Basal, HER2+** y agrupar los demás en **"Otros"**.  

#### **3. Tumor Other Histologic Subtype**
- Clasifica el tipo histológico del tumor en **Ductal NST, Tubular, Mucinous, etc.**  
- Algunos subtipos aparecen **menos de 50 veces** en un dataset de más de **2500 muestras**.  
- **Solución**: Mantener los más comunes y agrupar los restantes en **"Otros"**.  

#### **4. Oncotree Code**
- representa códigos de clasificación tumoral, los cuales pueden tener muchas categorías específicas. . 
- **Solución**: Para mejorar la calidad del modelo y evitar problemas, hemos decidido agrupar ciertas categorías dentro de la sección "Otros"


```python
umbral_frecuencia = 250

vars_a_reagrupar = [
    "Cancer Type Detailed", 
    "Pam50 + Claudin-low subtype", 
    "Tumor Other Histologic Subtype",
    "Oncotree Code"
]

def agrupar_categorias(df, columna, umbral):
    valores_frecuentes = df[columna].value_counts(dropna=False)  # Mantiene los valores previos imputados
    categorias_a_conservar = valores_frecuentes[valores_frecuentes >= umbral].index
    df[columna] = df[columna].apply(lambda x: x if x in categorias_a_conservar else "Otros")
    return df

for col in vars_a_reagrupar:
    df_cleaned = agrupar_categorias(df_cleaned, col, umbral_frecuencia)
print(df_cleaned["Cancer Type Detailed"].unique())
print(df_cleaned["Pam50 + Claudin-low subtype"].unique())
print(df_cleaned["Tumor Other Histologic Subtype"].unique())
print(df_cleaned["Oncotree Code"].unique())
```

### **6.1 Detección y eliminación de variables redundantes**

Según vimos anteriormente en el heatmap no hay ninguna varible numérica con una correlacion superior a 0.95 por lo que no tenemos variables con redundancia extrema 

```python
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Seleccionar solo variables numéricas para calcular la correlación
df_numeric = df_fe.select_dtypes(include=[np.number])

# Calcular la matriz de correlación
corr_matrix = df_numeric.corr()
```

```python
# Establecer un umbral de correlación alta
correlation_threshold = 0.95

# Identificar variables con correlación mayor al umbral
high_corr_vars = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > correlation_threshold:
            colname = corr_matrix.columns[i]
            high_corr_vars.add(colname)

# 📌 **Mostrar las variables con alta correlación**
print(f"Variables con correlación mayor a {correlation_threshold}: {high_corr_vars}")

# 📊 **Visualizar la matriz de correlación**
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="RdYlGn", fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlación entre Variables Numéricas")
plt.show()
```

```python

# Seleccionar solo las variables numéricas
numerical_features = df_numeric.select_dtypes(include=[np.number])

# Calcular VIF para cada variable
vif_data = pd.DataFrame()
vif_data["Variable"] = numerical_features.columns
vif_data["VIF"] = [variance_inflation_factor(numerical_features.values, i) for i in range(numerical_features.shape[1])]

# 📌 **Mostrar variables con VIF alto (indicador de redundancia)**
vif_data_sorted = vif_data.sort_values(by="VIF", ascending=False)
print("\nVariables con VIF alto (posible redundancia):")
print(vif_data_sorted)
```

### **6.4 Creación de variables derivadas**

#### - **Tumor Size Ratio** (Proporción del Tamaño del Tumor)
- Esta variable normaliza el tamaño del tumor para que los valores sean más comparables.
- Se resta la media y se divide por la desviación estándar → Estandarización Z-score.
- Permite detectar tumores significativamente más grandes o pequeños que el promedio.

Fórmula: Tumor Size Ratio=
Tumor Size−mean(Tumor Size)/
std(Tumor Size)

```python
df_fe["Tumor Size Ratio"] = (df_fe["Tumor Size"] - df_fe["Tumor Size"].mean()) / df_fe["Tumor Size"].std()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_fe["Tumor Size Ratio"], bins=30, color="pink", kde=True)

plt.title("Distribución de Tumor Size Ratio", fontsize=14)
plt.xlabel("Tumor Size Ratio")
plt.ylabel("Frecuencia")
```

#### - **Survival Ratio** (Ratio de Supervivencia Relativa)
- Compara los meses totales de supervivencia con el tiempo libre de recaída.
- Un ratio cercano a 1 indica que la paciente ha estado libre de recaída durante la mayor parte de su supervivencia.
- Un ratio mayor a 1 sugiere que la paciente vivió más tiempo después de la recaída.

Fórmula: Survival Ratio=
Overall Survival (Months)/Relapse Free Status (Months)+1

```python
df_fe["Survival Ratio"] = df_fe["Overall Survival (Months)"] / (df_fe["Relapse Free Status (Months)"] + 1)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_fe["Survival Ratio"], bins=30, color="pink", kde=True)

# Personalizar el gráfico
plt.title("Distribución de Survival Ratio", fontsize=14)
plt.xlabel("Survival Ratio")
plt.ylabel("Frecuencia")
```

#### - **Risk Index** (Índice de Riesgo Combinado basado en Nottingham Prognostic Index y Estadio Tumoral)
- Combina el Nottingham Prognostic Index (que evalúa el pronóstico del cáncer) con el estadio tumoral.
- Cuanto más alto, mayor el riesgo estimado.
- Puede ayudar a identificar pacientes con peor pronóstico

Fórumula: 
Risk Index = Nottingham prognostic index × Tumor Stage

```python
df_fe["Risk Index"] = df_fe["Nottingham prognostic index"] * df_fe["Tumor Stage"]
```

```python

plt.figure(figsize=(8, 5))
sns.countplot(x=df_fe["Risk Index"], color="pink")

# Personalizar el gráfico
plt.title("Distribución de Risk Index", fontsize=14)
plt.xlabel("Risk Index")
plt.ylabel("Frecuencia")
plt.xticks(rotation=45)  # Rotar etiquetas si hay muchos valores
plt.show()
```

#### - **Aggressive Treatment Score** 

- Cuantifica qué tan agresivo ha sido el tratamiento de la paciente.
- Mayor puntuación indica que la paciente ha recibido múltiples intervenciones.
- Basado en: Quimioterapia, Radioterapia, Terapia Hormonal y Tipo de Cirugía.

Fórmula: Aggressive Treatment Score=Chemotherapy+Radio Therapy+Hormone Therapy+Type of Breast Surgery

```python
df_fe["Aggressive Treatment Score"] = (
    df_fe["Chemotherapy"] + df_fe["Radio Therapy"] + df_fe["Hormone Therapy"] + df_fe["Type of Breast Surgery"]
)
```
